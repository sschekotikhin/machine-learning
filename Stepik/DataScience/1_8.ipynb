## Напишите программу. Тестируется через stdin → stdout

```python
my_stat['V5'] = my_stat.V1 + my_stat.V4
my_stat['V6'] = my_stat.V2.apply(math.log)

my_stat
```

## Напишите программу. Тестируется через stdin → stdout

```python
my_stat = my_stat.rename(
    columns={
        'V1': 'session_value',
        'V2': 'group',
        'V3': 'time',
        'V4': 'n_users'
    }
)

my_stat
```

## Напишите программу. Тестируется через stdin → stdout

```python
my_stat.session_value = my_stat.session_value.fillna(0)

n_users_median = my_stat.query('n_users >= 0').n_users.median()
my_stat.n_users = my_stat.n_users.apply(lambda x: n_users_median if x < 0 else x)

my_stat
```

## Напишите программу. Тестируется через stdin → stdout


```python
mean_session_value_data = my_stat.groupby('group', as_index=False) \
                                 .agg({'session_value': 'mean'}) \
                                 .rename(columns={'session_value': 'mean_session_value'})

mean_session_value_data
```
