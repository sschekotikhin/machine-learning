```python
import pandas as pd
import numpy as np
```

```python
sp_df = pd.read_csv('StudentsPerformance.csv')
sp_df.head()
```

```python
mean_scores = sp_df.groupby(['gender', 'lunch']) \
                   .aggregate({'math score': 'mean', 'reading score': 'mean'}) \
                   .rename(columns={'math score': 'mean_math_score', 'reading score': 'mean_reading_score'})
```

```python
mean_scores
```

```python
mean_scores.loc[[('female', 'standard'), ('female', 'free/reduced')]] # многоуровневые индексы
```

```python
sp_df.groupby(['gender']).lunch.unique()
```

```python
sp_df.sort_values(['gender', 'math score'], ascending=False).groupby('gender').head(10)
```

```python
total_score = sp_df.filter(like='score').sum(axis=1)
sp_df['total score'] = total_score
```

```python
sp_df
```

```python
sp_df = sp_df.assign(total_score_log=np.log(sp_df['total score']))
```

```python
sp_df
```

```python
sp_df.drop(labels=['total score', 'total_score_log'], axis=1)
```
