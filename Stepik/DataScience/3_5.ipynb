```python
import pandas as pd

import numpy as np

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix

import seaborn as sns
```

```python
mush_df = pd.read_csv('training_mush.csv')

X, y = mush_df.drop(labels=['class'], axis=1), mush_df['class']

mush_df
```

## Заполните пропуски

```python
clf = RandomForestClassifier(random_state=0)
```

```python
params = {
    'n_estimators': range(10, 50, 10),
    'max_depth': range(1, 12, 2),
    'min_samples_leaf': range(1, 7),
    'min_samples_split': range(2, 9, 2)
}

cv = GridSearchCV(clf, param_grid=params, cv=3, n_jobs=-1)
cv.fit(X, y)
```

```python
cv.best_params_
```

## Выберите один вариант из списка

odor > population > gill-attachment

```python
best_clf = cv.best_estimator_
```

```python
pd.DataFrame({
    'features': X.columns,
    'features_importances': best_clf.feature_importances_
}).sort_values(by=['features_importances'], ascending=False)
```

## Введите численный ответ

976

```python
test_mush_df = pd.read_csv('testing_mush.csv')

test_mush_df
```

```python
predictions = best_clf.predict(test_mush_df)

np.count_nonzero(predictions)
```

## Выберите один вариант из списка

1

```python
correct_y_test = pd.read_csv('testing_y_mush.csv')

confusion_matrix(correct_y_test, predictions)
```

```python
sns.heatmap(confusion_matrix(correct_y_test, predictions), cmap='Blues', annot=True)
```

```python
invasion_df = pd.read_csv('invasion.csv')

X_train, y_train = invasion_df.drop(labels=['class'], axis=1), invasion_df['class']

invasion_df
```

```python
clf = RandomForestClassifier(n_estimators=10)
```

```python
params = {
    'max_depth': range(2, 7, 2),
    'min_samples_leaf': range(1, 10, 1),
    'min_samples_split': range(2, 20, 2)
}

cv = GridSearchCV(clf, param_grid=params, cv=3, n_jobs=-1)
cv.fit(X=X_train, y=y_train)
```

```python
test_df = pd.read_csv('operative_information.csv')

test_df
```

```python
cv.best_params_
```

```python
best_clf = cv.best_estimator_

best_clf
```

```python
predictions = best_clf.predict(test_df)

for _class in np.unique(predictions):
    print(f'{_class}: {len(predictions[predictions == _class])}')
```
